```python
import mio
import pandas as pd
from matplotlib import pyplot as plt
```

```python
r00 = mio.read_raster('tile_670000_240000.bil')[0]
r10 = mio.read_raster('tile_680000_240000.bil')[0]
r01 = mio.read_raster('tile_670000_250000.bil')[0]
r11 = mio.read_raster('tile_680000_250000.bil')[0]
```

```python
l1 = pd.concat([r00, r10], axis=1)
```

```python
l2 = pd.concat([r01, r11], axis=1)
```

```python
l1.shape
```

```python
l2.shape
```

```python
res = (
    pd.concat([l1,l2])
    .reindex(
        columns = range(672500, 690000, 1),
        index = range(254000, 242000, -1)
    )
        
)
plt.imshow(res)
```

```python
mio.write_raster(res, '1091_veg.bil')
```
